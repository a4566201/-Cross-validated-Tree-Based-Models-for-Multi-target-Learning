```python
import time
from datasets import mse_comparison 
from scipy.io.arff import loadarff
import pandas as pd
import numpy as np
```

```python
raw_data = loadarff('datasets/scpf.arff')
df = pd.DataFrame(raw_data[0])
df.shape
```

```python
for i, col in enumerate(df.columns):
    print(i)
    print(df[col].isnull().sum())
    print(len(df[col].unique()))
    print(df[col].value_counts(normalize=True))
    print('-----------------------')
```

```python
df_new  = df.fillna(df.mode(axis=0).iloc[0])
```

```python
for i, col in enumerate(df_new.columns):
    print(f'Number of col:{i}')
    print(df_new[col].isnull().sum())
    print(len(df_new[col].unique()))
    print(df_new[col].value_counts(normalize=True))
    print('-----------------------')
```

```python
df_scaled = mse_comparison.scale_df(df=df_new,n_targets=3,scale_method='min_max')
```

```python
start = time.time()
_, _, _, mse_choose=  mse_comparison.mse_comparison(path=None,df=df_scaled, n_features=23,max_depth=np.inf,
                                 n_splits=10, only_choose_one=True,
                                  random_state=3)
print(f'Total runtime: {(time.time()-start)/60} minutes')
```

```python
start = time.time()
mse_all, mse_ind, mse_loo, _ = mse_comparison.mse_comparison(path=None,df=df_scaled, n_features=23,max_depth=np.inf,
                              use_pruning_for_loo=True,
                              use_pruning_2_models=True, n_splits=10,
                                  random_state=3)
print(f'Total runtime: {(time.time()-start)/60} minutes')
```

```python
import statistics
print(f'for all stdev {statistics.stdev(mse_all)}')
print(f'for ind stdev {statistics.stdev(mse_ind)}')
print(f'for loo stdev {statistics.stdev(mse_loo)}')
```
