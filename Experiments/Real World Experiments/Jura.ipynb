```python
df_scaled = mse_comparison.scale_df(df=df,n_targets=6,scale_method='min_max')
```

```python
start = time.time()
mse_all, mse_ind, mse_loo, mse_choose = mse_comparison.mse_comparison(path=None,df=df_scaled, n_features=15,max_depth=np.inf, use_pruning_for_loo=True,
                              use_pruning_2_models=True, n_splits=10,
                                  random_state=42)
print(f'Total runtime: {(time.time()-start)/60} minutes')
```

```python
import statistics
print(f'for all stdev {statistics.stdev(mse_all)}')
print(f'for ind stdev {statistics.stdev(mse_ind)}')
print(f'for loo stdev {statistics.stdev(mse_loo)}')
```
